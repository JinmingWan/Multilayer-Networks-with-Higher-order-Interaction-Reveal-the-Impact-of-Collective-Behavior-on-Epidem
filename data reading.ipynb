```python
import ast
import matplotlib.pyplot as plt
import pandas as pd
#baseline=pd.read_csv(".//result//HO_P_influence_result.csv")
def data_prevalence_plot(base,name=None):
    Status_count_1=base["node_count_1"]
    Status_count_2=base["node_count_2"]
    SC_res1=Status_count_1.values.tolist()
    SC_res2=Status_count_2.values.tolist()
    delta_f=ast.literal_eval
    Total1=delta_f(SC_res1[0])[0]+delta_f(SC_res1[0])[1]+delta_f(SC_res1[0])[2]+delta_f(SC_res1[0])[3]
    Total2=delta_f(SC_res2[0])[0]+delta_f(SC_res2[0])[1]+delta_f(SC_res2[0])[2]+delta_f(SC_res2[0])[3]    
    P_I_1=[]
    P_I_2=[]
    for result in SC_res1:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        P_I_1.append((E+I)/Total1)

    for result in SC_res2:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        P_I_2.append((E+I)/Total2)

    fig, ax = plt.subplots(figsize=(8,6))

    plt.plot(P_I_1, label="Community A", color="red")
    plt.plot(P_I_2, label="Community B", color="black")
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)

    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Prevalence z",fontsize=15)
    #plt.title("prevalence for SEIR Model simulation")
    plt.legend()
    image_format = 'eps' # e.g .png, .svg, etc.


    fig.savefig(name, format=image_format, dpi=1200)  
    plt.show()


        
        
        

def prevalence_plot(results):
    total1=5002
    total2=5002
    P_I_1=[]
    P_I_2=[]
    for result in results:
        P_I_1.append((result[0]["node_count_1"][1]+result[0]["node_count_1"][2])/total1)
        P_I_2.append((result[1]["node_count_2"][1]+result[1]["node_count_2"][2])/total2)
    plt.plot(P_I_1, label="Community A", color="red")
    plt.plot(P_I_2, label="Community B", color="black")
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)

    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Prevalence rate",fontsize=15)
    #plt.title("prevalence for SEIR Model simulation")
    plt.legend()
    plt.show()
    
def population_plot_percentage(results):
    S1=[]
    E1=[]
    I1=[]
    R1=[]
    S2=[]
    E2=[]
    I2=[]
    R2=[]
    total_1=5002
    total_2=5002
    for result in results:
        S1.append(result[0]["node_count_1"][0]/total_1)
        E1.append(result[0]["node_count_1"][2]/total_1)
        I1.append(result[0]["node_count_1"][1]/total_1)
        R1.append(result[0]["node_count_1"][3]/total_1)
        S2.append(result[1]["node_count_2"][0]/total_2)
        E2.append(result[1]["node_count_2"][2]/total_2)
        I2.append(result[1]["node_count_2"][1]/total_2)
        R2.append(result[1]["node_count_2"][3]/total_2)            
    plt.plot(S1, label="Susceptible", color="green")
    plt.plot(E1, label="Exposed", color="cyan")
    plt.plot(I1, label="Infectious", color="red")
    plt.plot(R1, label="Recoverd", color="blue")
    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Population fraction",fontsize=15)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15) 
    #plt.title("Layer 1 Population changes for SEIR Model simulation")
    plt.legend()
    plt.show()


    plt.plot(S2, label="Susceptible", color="green")
    plt.plot(E2, label="Exposed", color="cyan")
    plt.plot(I2, label="Infectious", color="red")
    plt.plot(R2, label="Recoverd", color="blue")
    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Population fraction",fontsize=15)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)    

    #plt.title("Layer 2 Population changes for SEIR Model simulation")
    plt.legend()
    plt.show()
    
def data_population_plot_percentage(base,location1=None,location2=None):
    S1=[]
    E1=[]
    I1=[]
    R1=[]
    S2=[]
    E2=[]
    I2=[]
    R2=[]
    Status_count_1=base["node_count_1"]
    Status_count_2=base["node_count_2"]
    SC_res1=Status_count_1.values.tolist()
    SC_res2=Status_count_2.values.tolist()
    delta_f=ast.literal_eval
    Total_1=delta_f(SC_res1[0])[0]+delta_f(SC_res1[0])[1]+delta_f(SC_res1[0])[2]+delta_f(SC_res1[0])[3]
    Total_2=delta_f(SC_res2[0])[0]+delta_f(SC_res2[0])[1]+delta_f(SC_res2[0])[2]+delta_f(SC_res2[0])[3]    

    for result in SC_res1:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        R=delta_f(result)[3]
        S1.append(S/Total_1)
        E1.append(E/Total_1)
        I1.append(I/Total_1)
        R1.append(R/Total_1)

    for result in SC_res2:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        R=delta_f(result)[3]
        S2.append(S/Total_2)
        E2.append(E/Total_2)
        I2.append(I/Total_2)
        R2.append(R/Total_2)
    
    
    fig, ax = plt.subplots(figsize=(8,6))
    plt.plot(S1, label="Susceptible", color="green")
    plt.plot(E1, label="Exposed", color="cyan")
    plt.plot(I1, label="Infectious", color="red")
    plt.plot(R1, label="Recoverd", color="blue")
    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Population fraction",fontsize=15)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15) 
    #plt.title("Layer 1 Population changes for SEIR Model simulation")
    plt.legend()
    image_format = 'eps' # e.g .png, .svg, etc.


    fig.savefig(location1, format=image_format, dpi=1200)  
    plt.show()
    
    
    
    
    
    fig, ax = plt.subplots(figsize=(8,6))
    plt.plot(S2, label="Susceptible", color="green")
    plt.plot(E2, label="Exposed", color="cyan")
    plt.plot(I2, label="Infectious", color="red")
    plt.plot(R2, label="Recoverd", color="blue")
    plt.xlabel("Time index t",fontsize=15)
    plt.ylabel("Population fraction",fontsize=15)
    plt.yticks(fontsize=15)
    plt.xticks(fontsize=15)    

    #plt.title("Layer 2 Population changes for SEIR Model simulation")
    plt.legend()
    
    
    image_format = 'eps' # e.g .png, .svg, etc.


    fig.savefig(location2, format=image_format, dpi=1200)   
    
  
    plt.show()

 
```

```python
def data_mat_converting(file1,name):
    base=pd.read_csv(file1)
    S1=[]
    E1=[]
    I1=[]
    R1=[]
    S2=[]
    E2=[]
    I2=[]
    R2=[]
    Status_count_1=base["node_count_1"]
    Status_count_2=base["node_count_2"]
    SC_res1=Status_count_1.values.tolist()
    SC_res2=Status_count_2.values.tolist()
    delta_f=ast.literal_eval
    Total_1=delta_f(SC_res1[0])[0]+delta_f(SC_res1[0])[1]+delta_f(SC_res1[0])[2]+delta_f(SC_res1[0])[3]
    Total_2=delta_f(SC_res2[0])[0]+delta_f(SC_res2[0])[1]+delta_f(SC_res2[0])[2]+delta_f(SC_res2[0])[3]    
    P_I_1=[]
    P_I_2=[]
    for result in SC_res1:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        R=delta_f(result)[3]
        S1.append(S/Total_1)
        E1.append(E/Total_1)
        I1.append(I/Total_1)
        R1.append(R/Total_1)
        P_I_1.append((E+I)/Total_1)

    for result in SC_res2:
        S=delta_f(result)[0]
        I=delta_f(result)[1]
        E=delta_f(result)[2]
        R=delta_f(result)[3]
        S2.append(S/Total_2)
        E2.append(E/Total_2)
        I2.append(I/Total_2)
        R2.append(R/Total_2)
        P_I_2.append((E+I)/Total_2)
    
       
    res={"S1":S1,"S2":S2,"E1":E1,"E2":E2,"I1":I1,"I2":I2,"R1":R1,"R2":R2,"P_I_1":P_I_1,"P_I_2":P_I_2}
    df=pd.DataFrame(data=res)
    df.to_csv(f".//Mat_read_results//{name}.csv")
    
```
